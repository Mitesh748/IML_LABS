```python
#to add, subtract, multiply, divide arguments element-wise 
import numpy as np
print("Add:")
print(np.add(5, 4))
print("Subtract:")
print(np.subtract(6, 4))
print("Multiply:")
print(np.multiply(2, 8))
print("Divide:")
print(np.divide(6, 3))
```

```python
#to round elements of the array to the nearest integer 
import numpy as np
x = np.array([2.8])
print("Original array:")
print(x)
x = np.rint(x)
print("Round elements of the array to the nearest integer:")
print(x)
```

```python
#to calculate mean across dimension, in a 2D numpy array 
import numpy as np
x = np.array([[20, 30], [30, 60]])
print("Original array:")
print(x)
print("Mean of each column:")
print(x.mean(axis=0))
print("Mean of each row:")
print(x.mean(axis=1))
```

```python
#to calculate the difference between neighboring elements, elementwise of a given array 
import numpy as np
x = np.array([0, 3, 5, 7, 9, 15])
print("Original array: ")
print(x)
print("Difference between neighboring elements")
print(np.diff(x))
```
